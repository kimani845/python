Data should be handled in the following ways
1. Drop data
2. Replace data
    1. Can be replaced with mean
    2. Can be replaced with median
    3. Can be replaced by its frequency
    4. Can be replaced based on the other functions



```python
# Deal with missing data
# Calculate the mean value of the normalized-loses column for each group
# grouped_mean = df.groupby('group')['normalized-losses'].mean()
# or can use this line of code for the same 
avr_norm_loss = df["normalized-losses"].astype('float').mean(axis=0)
print("Average of normalized-losses:", avr_norm_loss)

# replace NaN with the mean values in the column
df["normalized-losses"].replace(np.nan, avr_norm_loss, inplace=True)
```

```python
# calculate the mean value for the bore column 
avr_bore = df['bore'].astype('float').mean(axis=0)
print("Average bore:", avr_bore)

# Replace the mean of bpre with the value of bore
df['bore'].replace(np.nan, avr_bore, inplace=True)
```

```python
# calculating the mean value for the stroke column
avr_stroke = df['stroke'].astype('float').mean(axis=0)
print("Average for stroke:", avr_stroke)  # Output: 1.0

# replace the mean value for the stroke column
df['stroke'].replace(np.nan, avr_stroke, inplace=True)
```

```python
# calculate the mean value of the horse power column
avr_horse_power = df['horse-power'].astype('float').mean(axis=0)
print("Average for horse-power:", avr_horse_power)  # Output: 1.0

# replace the mean value for the horse power column
df['horse-power'].replace(np.nan, avr_horse_power, inplace=True)
```

```python
# Correcting the Data Format
# the last step in cleaning of data is checking and making sure that the data is in the correct format. in Pandas we use
#  .dtype() to check the dtatype and
#  .astype() to change the datatype

df.dtypes

```

```python
# in the above output, there are some of the columns that have incorrect data formats
# converting to correct data formats
df[['bore', 'stroke']] = df[['bore', 'stroke']].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[['peak-rpm']] = df[['peak-rpm']].astype("float")

# display the list after converting
df.dtypes

# we have successfully obtained cleaned data with no missing values and with data in its proper format
```

<h2 id="data_normalization">Data Normalization</h2>

<b>Why normalization?</b>
<p>Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include 
<ol>
    <li>scaling the variable so the variable average is 0</li>
    <li>scaling the variable so the variance is 1</li> 
    <li>scaling the variable so the variable values range from 0 to 1</li>
</ol>
</p>

<b>Example</b>
<p>To demonstrate normalization, say you want to scale the columns "length", "width" and "height".</p>
<p><b>Target:</b> normalize those variables so their value ranges from 0 to 1</p>
<p><b>Approach:</b> replace the original value by (original value)/(maximum value)</p>

```python
# replace the original value with by original value/maximum value
df ['length'] = df['length']/df['length'].max()
df ['width'] = df['width']/df['width'].max()
df ['height'] = df['height']/df['height'].max()
# show the scaled columns
# print(df[['length', 'width', 'height']])
df[['length', 'width', 'height']].head()
```

Binning is the process of transforming numerical variables into discrete categorical  'bins' for grouped analysis
For example, in the dataset, 'horse-power' is a real variable ranging from 48 to 288 and it has 59 unique values. what if you only care about the price difference between the cars with high horsepwer, medium horsepower , and little horsepower (3 types). You can rearrange them into three bins to simplify the analysis
Use the pandas method 'cut' to segment the horse-power  column into 3 bins.

```python
# # Replace non-numeric values with NaN
# df['horse-power'].replace('?', np.nan, inplace=True)

# # Convert the column to float first to handle NaN values
# df['horse-power'] = df['horse-power'].astype(float)

# Convert the column to int
df['horse-power'] = df['horse-power'].astype(int, copy=True)
# plot the histogram to see the distribution of horsepower
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import pyplot
plt.hist(df['horse-power'])
# set the x and y labels
plt.xlabel("horse-power")
plt.ylabel("count")
plt.title("Horse_power Bins")
```

```python
# finding 3 bins of equal bandwidth using numpy linspace(start_value, end_value, numbers_gemerated ) function. and since we want to include the minimum value of the horse power 
# start_value will be set to min(df["horse-power"])
# end_value will be set to max(df["horse-power"])
# and since we are using 3 bins of equal leengths, we will need 4 dividers, so 
# numbers_generated will be 4
bins = np.linspace(min(df["horse-power"]), max(df["horse-power"]), 4)
# set group names
group_names = ["Low", "Medium", "High"]
# apply the function cut to determine what each value of df['horse-power'] belongs to 
df['horse-power-binned'] = pd.cut(df['horse-power'], bins, labels=group_names, include_lowest= True)
df[['horse-power', 'horse-power-binned']].head(20)
```

```python
# see the number of vehicles in each bin
df['horse-power-binned'].value_counts()
```

```python
# plot the distribution of each bin
pyplot.bar(group_names, df['horse-power-binned'].value_counts())

# set the x/y labels
plt.xlabel('Horse_power')
plt.ylabel('Count')
plt.title('Horsepower Bins')

# i have narrowed down the intervals from 59 to 3
```

```python
# bins visualization
# normally we use histogram to visualize the bins that we have created above
# draw the histogram of the attribute 'horsepower' with 3 bins 

plt.hist(df['horse-power'], bins=3)
plt.xlabel("Horsepoer")
plt.ylabel("Frequency")
plt.title('horsepoer bins')
```

INDICATOR VARIABLE OR THE DUMMY VARIABLE <br>
An indicator vsriable is a numerical variable used to label categories. they are called dummies because the numbers themselves doesnt have inherent meaning
Indicator variables are used so that you can use categorical variables for regression analysis . They are used to convert categorical data into numerical data.
In our dataset, we have fuel-type which has two variables. 'gas' or 'diesel' regression doesnt understand words, only numbers. to use this attribute in regression analysis, you can 
convert the 'fuel type' into indicator variables.


```python
df.columns
```

```python
# get the indicator variables and assign it to dummy variables
dummy_variable_1 = pd.get_dummies(df['fuel_type'])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel': 'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# merge the dataframe 'df' and 'dummy_variable_1'
df = pd.concat([df, dummy_variable_1], axis=1)

# drop the original column 'fuel-type' from df
df.drop('fuel_type', axis=1, inplace=True)
df.head()
```
